```python
import tensorflow.keras
import segmentation_models as sm
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import array_to_img
from tensorflow.keras.models import load_model

import os
import cv2
import numpy as np
import random
```

```python
import tensorflow.keras
# or from tensorflow import keras

tensorflow.keras.backend.set_image_data_format('channels_last')
# or keras.backend.set_image_data_format('channels_first')
```

```python
# Load pre trained PSPNet segmentation model

model = sm.PSPNet()
```

```python
# Set the model parameters

model = sm.PSPNet('resnet34', classes=1, activation='sigmoid', encoder_weights='imagenet', input_shape=(240, 240, 3))
```

```python
# Change the dataset directory in which photos + mask of a location are stored
directory = "1093"
```

```python
dim = 0

for root, dirs, files in os.walk(directory + "/"):
    for file in files:
        if file.endswith(".jpg"):
            dim=dim+1
        
print("dim: ", dim)

# change the first value(300 here) to the number of images to sample from the folder 
images_array = np.empty([300, 57600, 3], dtype="float")

i=0;


for root, dirs, files in os.walk(directory + "/"):
    for file in files:
        if file.endswith(".jpg"):
            img_path = os.path.join(root, file)
            img = cv2.imread(img_path)
            # image resolution is downscaled to 240 * 240 pixels
            res = cv2.resize(img, dsize=(240, 240), interpolation=cv2.INTER_CUBIC)
            img_array = img_to_array(res)
            # 57600 = 240*240 and 3 represents the number of color channels
            img_array = img_array.reshape(57600, 3)
            images_array[i] = img_array
            print("\n\nImages: ", images_array[i], "\n", i, " : ", img_array)
            i=i+1
            print("i", i)
            # i=300 means it'll take the first 300 images' data which will later be assignes as 240 for training and 60 as testing
            if(i==300):
                break
```

```python
# print array sample
images_array[69]
```

```python
# add the mask data as the prediction class
img_path = directory + "/" + directory + "_mask.png"
img = cv2.imread(img_path)
res = cv2.resize(img, dsize=(240, 240), interpolation=cv2.INTER_CUBIC)
gray = cv2.cvtColor(res, cv2.COLOR_BGR2GRAY)

img_array = img_to_array(gray)
# print(img_array)
print(img_array.shape)
predict_array = img_array.reshape(57600, 1)
print(predict_array.shape)
```

```python
# assign 240 images for training and 60 for testing

y_train = np.empty([240, 57600, 1], dtype="float")

for i in range (0, 240):
    y_train[i]=predict_array


y_test = np.empty([60,57600, 1], dtype="float")

for i in range (0, 60):
    y_test[i]=predict_array
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
X_test = np.empty([60, 57600, 3], dtype="float")
X_train = np.empty([240, 57600, 3], dtype="float")

j=0
k=0

for i in range (0, 300):
    if(i%5==0):
        X_test[j] = images_array[i]
        j=j+1
    else:
        X_train[k] = images_array[i]
        k=k+1
```

```python
# Normalize the pixel values
X_test/=255
X_train/=255
y_test/=255
y_train/=255
```

```python
# reshape the data
X_train = X_train.reshape(240,240,240,3)
X_test = X_test.reshape(60,240,240,3)
y_train = y_train.reshape(240,240,240,1)
y_test = y_test.reshape(60,240,240,1)
```

```python
# pre-process images data for the segmentation model
BACKBONE = 'resnet34'
preprocess_input = sm.get_preprocessing(BACKBONE)

x_train, y_train, x_val, y_val = X_train, y_train, X_test, y_test

x_train = preprocess_input(x_train)
x_val = preprocess_input(x_val)
```

```python
# compile the model

model.compile(
    'Adam',
    loss=sm.losses.bce_jaccard_loss,
    metrics=[sm.metrics.iou_score],
)
```

```python
# train the model
model.fit(
   x=x_train,
   y=y_train,
   batch_size=4,
   epochs=2,
   validation_data=(x_val, y_val),
)
```

```python
model.save("skyPixelModel.h5")
```
